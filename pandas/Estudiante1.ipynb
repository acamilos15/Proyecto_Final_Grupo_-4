```python
import pandas as pd
```

```python
df = pd.read_csv("Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv")
df.head()
```

```python
def sumar_digitos(x):
    suma = 0
    for c in str(x):
        suma += int(c)
    return suma
def funcion(n):
    def f(x):
        return n*(x**2) + 2*x + 23
    return f

def evaluar_punto2(n):
    f = funcion(n)
    return round((f(101) - f(100)) / 1, 2)

df["n"] = df["Número de documento"].apply(sumar_digitos)
df["Punto 2"] = df["n"].apply(evaluar_punto2)


p2 = df.columns[2]

def calificar2(x):
    return 5 if x[p2] == x["Punto 2"] else 1

df["Calificacion 2"] = df.apply(calificar2, axis=1)

```

```python
def cam(x):
  p=str(x)
  p=p.replace(",",".")
  return float(p)

def sumar_digitos(x):
    x_str = str(x)
    sum=0
    for i in x_str:
      sum+= int(i)
    return sum

def crear_funcion1(n):
    def f(x):
        return n * (2*x) + 2
    return f

def evaluar_funcion3(n):
  f=crear_funcion1(n)
  return f(100)


df["n"]= df["Número de documento"].apply(sumar_digitos)
df["Punto 3"] = df["n"].apply(evaluar_funcion3).round(2)
columna_estudiante = df.columns[3]
df["Punto 3 Limpio"] = df[columna_estudiante].apply(cam)



def calificar3(x):
  return 5 if x["Punto 3"] == x["Punto 3 Limpio"] else 1

df["Calificacion 3"] = df.apply(calificar3 , axis=1)
```
