```python
import pandas as pd
```

```python
df = pd.read_csv("Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv")
df.head()
```

```python
def sumar_digitos(x):
    x_str = str(x)
    sum=0
    for i in x_str:
      sum+= int(i)
    return sum

def funcion(n):
  def f(x):
    return n*(x**2)+2*x+23
  return f

def evaluar(n):
  f=funcion(n)
  return (f(101)-f(100))/5
df["n"]= df["Número de documento"].apply(sumar_digitos)
df["Punto 2"] = df["n"].apply(evaluar).round(2)

def calificar2(x):
  return 5 if x[1] == x["Punto 2"] else 1

df["Calificacion 2"] = df.apply(calificar2 , axis=1)
```
