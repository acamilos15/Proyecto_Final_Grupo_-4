```python
df = pd.read_csv("Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv")
df.head()
```

```python
def sumar_digitos(x):
    suma = 0
    for c in str(x):
        suma += int(c)
    return suma
def funcion(n):
    def f(x):
        return n*(x**2) + 2*x + 23
    return f

def evaluar_punto2(n):
    f = funcion(n)
    return round((f(101) - f(100)) / 1, 2)

df["n"] = df["Número de documento"].apply(sumar_digitos)
df["Punto 2"] = df["n"].apply(evaluar_punto2)


p2 = df.columns[2]

def calificar2(x):
    return 5 if x[p2] == x["Punto 2"] else 1

df["Calificacion 2"] = df.apply(calificar2, axis=1)

```

```python
def cam(x):
  p=str(x)
  p=p.replace(",",".")
  return float(p)

def sumar_digitos(x):
    x_str = str(x)
    sum=0
    for i in x_str:
      sum+= int(i)
    return sum

def crear_funcion1(n):
    def f(x):
        return n * (2*x) + 2
    return f

def evaluar_funcion3(n):
  f=crear_funcion1(n)
  return f(100)


df["n"]= df["Número de documento"].apply(sumar_digitos)
df["Punto 3"] = df["n"].apply(evaluar_funcion3).round(2)
columna_estudiante = df.columns[3]
df["Punto 3 Limpio"] = df[columna_estudiante].apply(cam)



def calificar3(x):
  return 5 if x["Punto 3"] == x["Punto 3 Limpio"] else 1

df["Calificacion 3"] = df.apply(calificar3 , axis=1)
```

```python

def desviacion(archivo, columna):
    try:
        df = pd.read_csv(archivo) 
        return df[columna].std()
    except Exception as e:
        print("Error al procesar la columna:", e)
        return None
```

```python
import os
import matplotlib.pyplot as plt
from funciones import desviacion
import pandas as pd

NOMBRE_ARCHIVO = "Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv" 

# 2. Asegúrate de que estos nombres de columna sean CORRECTOS y existan en tu archivo CSV
COLUMNAS_A_MEDIR = ['CP1','CP2','CP3', 'NF'] 

# 3. Ruta donde se guardará la imagen
RUTA_CARPETA = "figuras"
NOMBRE_GRAFICO = 'desviacion_estandar_resultados.png'
RUTA_COMPLETA = os.path.join(RUTA_CARPETA, NOMBRE_GRAFICO)

# 1. Calcular desviaciones para todas las columnas
print("\nCalculando desvíos...")
resultados_desvios = {}

for columna in COLUMNAS_A_MEDIR:
    std_dev = desviacion(NOMBRE_ARCHIVO, columna)
    resultados_desvios[columna] = std_dev
    print(f"Desviación de '{columna}': {std_dev}")

# 2. Convertir los resultados en una Serie de Pandas para graficar
# Esto es crucial para que el método .plot() de Pandas funcione o para pasar los datos a Matplotlib
df_grafico = pd.Series(resultados_desvios)

# --- 3. Filtrar los valores NaN (donde el cálculo falló) antes de graficar ---
# Esto evita que los valores 'nan' causen problemas en la gráfica.
df_grafico = df_grafico.dropna() 

# --- 4. Verificación y Graficación ---
if df_grafico.empty:
    print("\n⚠️ ERROR: No hay datos numéricos válidos para graficar después de calcular las desviaciones.")
else:
    print("\nGenerando gráfico...")
    
    # Crea el gráfico de barras usando Pandas o Matplotlib
    # Uso Matplotlib como en tu código original, pero usando la Serie 'df_grafico'
    fig, ax = plt.subplots(figsize=(8, 6)) # Define el tamaño del gráfico
    
    # Gráfico de barras usando los índices (nombres de columna) y los valores (desviaciones)
    colores = ['skyblue', 'lightcoral', 'lightgreen', 'gold']
    barras = ax.bar(df_grafico.index, df_grafico.values, color=colores[:len(df_grafico.index)])
    
    # Configuración del gráfico (como en tu código)
    plt.title('Desviación Estándar de Puntuaciones por Columna', fontsize=14, fontweight='bold')
    plt.xlabel('Columnas', fontsize=12)
    plt.ylabel('Desviación Estándar', fontsize=12)
    plt.xticks(rotation=45, ha='right') # Rotar etiquetas para mejor visualización
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    
    # Mostrar los valores encima de las barras
    for bar in barras:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval + 0.05, f'{yval:.2f}', 
                 ha='center', va='bottom', fontsize=10)

    plt.tight_layout() # Ajusta el layout para evitar solapamiento

    # Guardar la Figura
    # --- Crear la carpeta 'figuras' si no existe ---
    if not os.path.exists(RUTA_CARPETA):
        os.makedirs(RUTA_CARPETA)
        print(f"Carpeta '{RUTA_CARPETA}' creada.")

    # --- Guardar la figura ---
    plt.savefig(RUTA_COMPLETA)
    print(f"✅ Diagrama de barras guardado en: {RUTA_COMPLETA}")

    # Mostrar la Figura (opcional, dependiendo del entorno)
    plt.show()
    
```
