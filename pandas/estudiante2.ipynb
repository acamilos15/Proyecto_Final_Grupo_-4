```python
import pandas as pd

df = pd.read_csv("Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv")

# Calcular Nota Final (promedio de los 3 puntos)
df["Calificacion Final"] = ((df["Calificacion 1"] + df["Calificacion 2"] + df["Calificacion 3"]) / 3).round(2)

# Guardar
df.to_csv("Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv", index=False)


print(df[["Número de documento", "Calificacion 1", "Calificacion 2", "Calificacion 3", "Calificacion Final"]].head())
```
