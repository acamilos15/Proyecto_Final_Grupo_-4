```python
import pandas as pd

def percentiles(nombre_archivo, nombre_columna):
    try:
        
        datos = pd.read_csv(nombre_archivo)
        
        
        if nombre_columna not in datos.columns:
            return "No existe la comuna"
        
        
        p25 = datos[nombre_columna].quantile(0.25)
        p50 = datos[nombre_columna].quantile(0.50)
        p75 = datos[nombre_columna].quantile(0.75)
        
        # Retornar en el mismo formato que trabajaron
        return {
            'percentil_25': p25,
            'percentil_50': p50,
            'percentil_75': p75
        }
        
    except:
        return "No existe la comuna"
```
