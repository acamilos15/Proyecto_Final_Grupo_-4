```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# 1. Importamos la función (asumiendo que funciones.py está en la misma carpeta 'pandas')
from funciones import percentiles

# --- CONFIGURACIÓN ---
nombre_archivo = "Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv" # CAMBIA ESTO por el nombre real de tu csv
columna_objetivo = 'NF'

# Solo para probar que la importación funciona (parte del requisito)
resultado_funcion = percentiles(nombre_archivo, columna_objetivo)
print(f"Prueba de función importada: {resultado_funcion}")

# --- GENERACIÓN DE LA GRÁFICA ---

# Leemos los datos para generar los percentiles 10-90
try:
    df = pd.read_csv(nombre_archivo)
    
    if columna_objetivo in df.columns:
        # Generamos la lista de percentiles del 10 al 90 (0.1 a 0.9)
        rangos = [x/100 for x in range(10, 100, 10)] # [0.1, 0.2, ..., 0.9]
        valores = df[columna_objetivo].quantile(rangos)
        
        # Crear la figura
        plt.figure(figsize=(10, 6))
        
        # Crear gráfico de barras
        # Usamos los índices (0.1, 0.2...) como etiquetas x, y los valores como altura
        barras = plt.bar([f"{int(x*100)}%" for x in rangos], valores, color='skyblue', edgecolor='black')
        
        plt.title('Percentiles')
        plt.xlabel('Percentil')
        plt.ylabel(f'Valor de {columna_objetivo}')
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        
        # Añadir los valores encima de las barras (opcional, pero se ve mejor)
        for barra in barras:
            height = barra.get_height()
            plt.text(barra.get_x() + barra.get_width()/2., height,
                     f'{height:.2f}', ha='center', va='bottom')

        # --- GUARDAR EN LA CARPETA FIGURAS ---
        # Dado que estás en la carpeta 'pandas', la ruta relativa es 'figuras/nombre.png'
        ruta_guardado = os.path.join('figuras', 'Percentiles.png')
        
        # Verificamos que la carpeta exista, si no, la crea (por seguridad)
        os.makedirs('figuras', exist_ok=True)
        
        plt.savefig(ruta_guardado)
        print(f"Gráfica guardada exitosamente en: {ruta_guardado}")
        
        plt.show()
        
    else:
        print("Error: La columna 'NF' no existe en el DataFrame para graficar.")

except FileNotFoundError:
    print(f"Error: No se encontró el archivo {nombre_archivo}")
```
