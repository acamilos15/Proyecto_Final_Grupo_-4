```python
import pandas as pd
```

```python
df = pd.read_csv("Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv")
df.head()
```

```python
def sumar_digitos(x):
    x_str = str(x)
    sum=0
    for i in x_str:
      sum+= int(i)
    return sum

def crear_funcion(n):
    def f(x):
        return n * (x**2) + 2*x + 23
    return f

def evaluar_funcion1(n):
  f=crear_funcion(n)
  return (f(105)-f(100))/5

df["n"]= df["Número de documento"].apply(sumar_digitos)
df["Punto 1"] = df["n"].apply(evaluar_funcion1).round(2)

def calificar1(x):
  return 5 if x[1] == x["Punto 1"] else 1

df["Calificacion 1"] = df.apply(calificar1 , axis=1)

```
