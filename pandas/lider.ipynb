```python
def sumar_digitos(x):
    x_str = str(x)
    sum=0
    for i in x_str:
      sum+= int(i)
    return sum

def crear_funcion(n):
    def f(x):
        return n * (x**2) + 2*x + 23
    return f

def evaluar_funcion1(n):
  f=crear_funcion(n)
  return (f(105)-f(100))/5

df["n"]= df["NÃºmero de documento"].apply(sumar_digitos)
df["Punto 1"] = df["n"].apply(evaluar_funcion1).round(2)

def calificar1(x):
  return 5 if x[1] == x["Punto 1"] else 1

df["Calificacion 1"] = df.apply(calificar1 , axis=1)

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# 1. Importar la funciÃ³n 'promedios' desde el archivo 'funciones.py'
# AsegÃºrate de que 'funciones.py' contenga la definiciÃ³n de la funciÃ³n 'promedios'.
from funciones import promedios

## --- ConfiguraciÃ³n de Archivos y Columnas ---
NOMBRE_CSV = "Quiz CÃ¡lculo 1- Gobierno 22 de Octubre(1-18).csv"
COLUMNAS = ["CP1", "CP2", "CP3", "NF"]
promedios_calculados = {}

# 2. Calcular los promedios utilizando la funciÃ³n importada
print("ðŸ“Š Calculando promedios...")
for columna in COLUMNAS:
    # Llama a la funciÃ³n 'promedios' para obtener el valor
    promedio = promedios(NOMBRE_CSV, columna)
    if promedio is not None:
        promedios_calculados[columna] = promedio
        print(f"El promedio de {columna} es: {promedio:.2f}")

# 3. Preparar los datos y definir la figura
datos_grafico = pd.Series(promedios_calculados)

plt.figure(figsize=(8, 6)) # Define el tamaÃ±o del grÃ¡fico
# Crear el grÃ¡fico de barras
datos_grafico.plot(kind='bar', color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])

# 4. ConfiguraciÃ³n del grÃ¡fico
plt.title('Promedio de Puntuaciones por Columna', fontsize=14, fontweight='bold')
plt.xlabel('Columnas', fontsize=12)
plt.ylabel('Valor Promedio', fontsize=12)
plt.xticks(rotation=0) # Etiquetas horizontales
plt.grid(axis='y', linestyle='--', alpha=0.7)
# Mostrar los valores encima de las barras
for i, valor in enumerate(datos_grafico):
    plt.text(i, valor + 0.05, f'{valor:.2f}', ha='center', va='bottom', fontsize=10)

plt.tight_layout()

## --- Guardar la Figura ---
# 5. Definir la ruta de guardado
RUTA_CARPETA = 'figuras'
NOMBRE_ARCHIVO = 'promedios_quiz.png'
RUTA_COMPLETA = os.path.join(RUTA_CARPETA, NOMBRE_ARCHIVO)

# 6. Crear la carpeta 'figuras' si no existe
if not os.path.exists(RUTA_CARPETA):
    os.makedirs(RUTA_CARPETA)
    print(f"\nðŸ“ Carpeta '{RUTA_CARPETA}' creada.")

# 7. Guardar la figura
# Se guarda la figura en la ruta especificada.
plt.savefig(RUTA_COMPLETA)
print(f"âœ… Diagrama de barras guardado en: pandas/{RUTA_COMPLETA}")

# 8. Mostrar la figura
plt.show()
```
