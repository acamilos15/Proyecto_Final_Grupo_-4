```python
import pandas as pd

df = pd.read_csv('test.csv')

conteo_por_clase = df['Pclass'].value_counts().sort_index()

edad_promedio_por_sexo = df.groupby('Sex')['Age'].mean()


print("--- Conteo de Pasajeros por Clase (Pclass) ---")
print(conteo_por_clase)
print("\n--- Edad Promedio (Age) por Sexo (Sex) ---")
print(edad_promedio_por_sexo)
```
